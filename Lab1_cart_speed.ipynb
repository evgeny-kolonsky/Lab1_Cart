# Introduction

In this experiment, we will examine the advantages and disadvantages of direct versus indirect measurement in terms of the accuracy they allow. In the direct measurement, we will examine and quantify the effect of the random error on the results of the experiment, by repeating the measurement many times.

We want to measure the average speed of a cart in a sloping plane. This measurement can only be done indirectly: the distance traveled by the cart must be measured and divided by the duration of the movement. The duration can be measured directly using a detector and indirectly from the solution of the forces equation and measurement of the geometric parameters of the system.

The experimental system appears in Figure1 and consists of a trolley moving on an air rail and a photoelectric gate that measures the length of time ($\Delta t$[sec])  that the cart passes through. The air rail minimizes friction in the system so that its effect is negligible. The parameters that cannot be changed are $l$ [m]- the length of the wagon and $L$ [m]- the length of the air rail. $S$[m]  and the rail elevation $h$[m] that creates the slope can be changed as needed.

# Theory

### Theoretical value $\Delta t$

$$\Delta t = \sqrt{2sL\over gh} \left( \sqrt{1+l/s} - 1\right)  $$

```python
dt_theor = np.sqrt(2* s * L /  g / h) * (np.sqrt(1+l/s) - 1)
dt_theor
```

### Uncertainty

Denote $\delta x$ as the relative uncertainty of value $x$: $\delta x = {\tilde x \over x}$, where $\tilde x$ is absolute uncertainty of $x$.

$$
\delta{\Delta t} = {1\over 2}\sqrt{(\delta s^2 + \delta L^2 + \delta g^2 + \delta h^2 + \delta l^2)}
$$

```python
unc_dt = np.sqrt((ds/s)**2 + (dL/L)**2 + (dg/g)**2 + (dh/h)**2 + (dl/l)**2) /2
print(f'Theory Delta t: {dt_theor:.2f} ±{dt_theor * unc_dt:.2f} c')
```

### Chauvenet's criterion

$$
N  P( |x| > k) < {1\over 2}, \quad k = {|x_{outlier} - \hat x| \over \sigma}
$$

```python
from scipy.stats import norm

mu = t.mean()
sigma = t.std()
t_outlier = np.max(t)

k = np.abs(t_outlier - mu)/sigma
P = (1 - norm.cdf(k)) + norm.cdf(-k)
N = t.size
print(f'Chauvene criterion value to be compared with 1/2: {N*P:.1e}')
if N * P < .5:
  print(f'value {t_outlier:.3f} is outlier: to be deleted')
  t = np.delete(t, np.argmax(t))
  # new average and sigma
  mu = t.mean()
  sigma = t.std()
  N = t.size
```

```python
def gauss(t, mu, sigma):
  return np.exp(-(t-mu)**2/2/sigma**2) / np.sqrt(2*np.pi*sigma**2)

bins = 10
tspace = np.linspace(mu-3*sigma, mu+3*sigma, 100)
plt.plot(tspace, gauss(tspace, mu, sigma), linewidth=3)

#plt.hist(t, bins=13, density=True, alpha=0.5,edgecolor='gray',color='steelblue', histtype='bar'); #normalized
plt.hist(t, bins=10, density=True, alpha=0.1,color='steelblue', edgecolor='gray', histtype='bar'); #normalized
ymax = gauss(mu-sigma/2, mu, sigma)
#plt.vlines([mu-sigma/2, mu+sigma/2], ymin=0,  ymax= ymax, color='red', alpha=0.5)
#plt.hlines( ymax, xmin=mu-sigma/2, xmax=mu+sigma/2, color='red', alpha=0.5)
plt.axis('off')
```
